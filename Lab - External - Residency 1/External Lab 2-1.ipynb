### Find out the no of male & female employees in the company

*Hint: Use can use a condition to find out the no of 1s and 0s in gender and sum it*

```python
# Unique () for the gender column shows 1's 2's  rather than 1's & 0's 
# Hence assuming 1's = Male , 2's = Female 
hr.Gender.unique()
# returns series 
hr['Gender'].value_counts()
```

```python
male = hr['Gender'].value_counts().get(1)
female = hr['Gender'].value_counts().get(2)
print('The number of male employees in the company : {0} \nThe number of female employees in the company : {1}'.format(male,female))

# Bar plot for the employee distribution. 

gender_plot = pd.DataFrame({'Gender':['Male', 'Female'], 'Count':[male, female]})
gender_plot.plot.bar(x='Gender', y='Count', rot=0)
```

### Find out the IQR for the column - Age 

*Hint: IQR - Interquartile range = 75th percentile - 25th percentile* 

```python
percentile_75 = hr['Age'].quantile(0.75)
percentile_25 = hr['Age'].quantile(0.25)
IQR = percentile_75 - percentile_25
print('The Interquartile range for the column "Age" is : {0}'.format(IQR))
```

### Find out the mean, median and variance for the column - DailyRate 

```python
mean = hr['DailyRate'].mean()
median = hr['DailyRate'].median()
variance = hr['DailyRate'].var()
print('The column DailyRate values,  \nMean : {0} \nMedian : {1} \nVariance : {2} '.format(mean,median,variance))
```

### Create 500 random samples from normal distribution with mean 1 and  standard deviation 0.5

```python
mu = 1 #mean
sigma = 0.5 #standard deviation
ndata = np.random.normal(mu, sigma, 500)
display(ndata)
```

###  Plot the histogram of the above normal distribution 

```python
#using seaborn for ploting the normal distribution and density curve. 

ndist = sns.distplot(ndata, hist=True, kde=True, label='Normal Dist \nMean : {0} \nStd. Dev : {1}'.format(mu,sigma),
             bins=50, color = 'green', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4 , 'color' : 'red'})
plt.legend()
```
